### Q1. What is the overall sales trend?

```python
## to plot on graph we need to change period to object type:
data['Month_year']=data['Month_year'].astype(str)
```

```python
data.dtypes
```

```python
Month_year=data.groupby(['Month_year'])['sales'].sum()
Month_year
```

```python
plt.figure(figsize=(15,15))
plt.plot(Month_year.keys(),Month_year)
plt.xticks(Month_year.keys(),rotation='vertical',size=15);
```

 
### Q2. Which are the Top 10 products by sales?


```python
top10_products=data.groupby(['product_name'])['sales'].sum().to_frame()
```

```python
top10_products.sort_values('sales',ascending=False).head(10)
```

### Q3. Which are the Most Selling Products?

```python
most_sell_products=data.groupby(['product_name'])['quantity'].sum().to_frame()
most_sell_products=most_sell_products.sort_values('quantity',ascending=False)
most_sell_products
```
