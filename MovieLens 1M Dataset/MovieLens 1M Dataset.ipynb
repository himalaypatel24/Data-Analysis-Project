### Loading Data into DataFrame:

```python

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('C:\\Users\\TEMP\\ratings.dat.txt', sep='::',
 header=None, names=rnames)
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('C:\\Users\\TEMP\\movies.dat.txt', sep='::',
 header=None, names=mnames)
```

### Using pandasâ€™s merge function, we first merge ratings with users and then merge that result with the movies data. pandas infers which columns to use as the merge (or join) keys based on overlapping names:

```python
all_data= (pd.merge(pd.merge(ratings, users), movies))
```

```python
all_data
```

### Lets get top 20 movieon basis on mean ratings for each film:

```python
top_20=all_data.groupby(['title'])['rating'].mean()
top_20=top_20.sort_values(ascending=False).head(20)
top_20
```

### Lets get mean movie ratings for each film grouped by gender, we can use the
### pivot_table method:

```python
all_data.pivot_table('rating',index='title',columns='gender',aggfunc='mean')
```

```python
means =all_data.groupby(['title','gender'])['rating'].mean().unstack()
means
```

### filter down to movies that received at least 250 ratings (a completely arbitrary number); to do this,

```python
ratings_by_title = all_data.groupby('title').size()
ratings_by_title.head()
```

```python
titles=all_data.groupby(['title'])['rating'].count()
```

```python
titles=titles.keys()[titles>250]
titles
```

```python
mean_ratings=means.loc[titles]
mean_ratings
```

### To see the top films among female and male viewers:

```python
female=mean_ratings.sort_values(by='F',ascending=False).head(10)
female
```

```python
male=mean_ratings.sort_values(by='M',ascending=False).head(10)
male
```

### Lets find common movie popular in male and female:

```python
common =pd.merge(male,female,how='inner',on='title')
```

```python
common
```

### Measuring Rating Disagreement

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
 sorted_by_diff=mean_ratings.sort_values('diff',ascending=True)
sorted_by_diff[::-1][:10]
```
